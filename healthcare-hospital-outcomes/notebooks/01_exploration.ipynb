# Hospital Outcomes & Quality — Exploratory Analysis

This notebook explores hospital-level quality metrics using CMS public data.
The goal is to understand distribution, trends, and data quality before building KPIs and dashboards.


```python
import pandas as pd

df = pd.read_csv("C:/Users/spand/Downloads/cloud-analytics-portfolio-main/cloud-analytics-portfolio-main/healthcare-hospital-outcomes/data/sample/Hospital_General_Information.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.info()
```

```python
df["hospital_overall_rating"] = pd.to_numeric(df["hospital_overall_rating"], errors="coerce")
```

```python
df.groupby("state")["hospital_overall_rating"].mean().sort_values(ascending=False)
```

```python
df["hospital_overall_rating"].isna().sum()
```

```python
# Export curated dataset for Tableau
tableau_df = (
    df.groupby("state", as_index=False)
      .agg(
          hospital_count=("facility_name", "count"),
          avg_rating=("hospital_overall_rating", "mean"),
          missing_ratings=("hospital_overall_rating", lambda x: x.isna().sum())
      )
)

tableau_df.to_csv("../data/curated/state_quality_kpis.csv", index=False)
```
